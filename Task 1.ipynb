**Visualization 1: Distribution of Genders (Bar Plot)**

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('adult.csv')

# Display the first few rows of the dataset 
print(df.head())

# Set the style of the visualization
sns.set(style="whitegrid")

# Plotting a bar plot of Gender
plt.figure(figsize=(8, 6))
gender_counts = df['gender'].value_counts()
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette=['cyan', 'purple'])

# Add titles and labels
plt.title('Distribution of Genders')
plt.xlabel('Gender')
plt.ylabel('Count')

# Show the plot
plt.show()
```

**Visualization 2: Distribution of Ages (Histogram)**

```python
# Set the style of the visualization
sns.set(style="whitegrid")

# Plotting a histogram of Age 
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=30, kde=True)

# Add titles and labels
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
